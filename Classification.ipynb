```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/Users/ASUS/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors = 'coerce')
```

```python
df.info()
```

```python
df.dropna(how = 'any', inplace = True)
```

```python
df.head()
```

```python
df.Churn.value_counts()/len(df)*100
```

```python
x = df.drop(['customerID','Churn'], axis = 1)
y = df.Churn.values
```

```python
x.columns
```

```python
# convert catogorical features into numerical

x = pd.get_dummies(x, columns=['gender', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],drop_first=True, dtype = int)
```

```python
x.head(1)
```

```python
#splitting data

from sklearn.model_selection import train_test_split
x_train ,x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
len(x_train), len(x_test)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x_train_sc
```

```python
x_test
```

# KNN

```python
#KNN classifier

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

model.fit(x_train_sc, y_train)
```

```python
y_pred = model.predict(x_test_sc)
```

```python
y_pred,y_test
```

```python
# classification matrix

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)*100
```

```python
x_test.head(1)
```

```python
data_sc = sc.transform(data)

ND = model.predict(data_sc)

print(ND)
```

## Question 2: Data Cleaning and Preprocessing
Based on the missing values, decide how to handle them for 'Age', 'Cabin', and 'Embarked'.<br>
For 'Age', fill missing values with the median age.<br>
For 'Embarked', fill missing values with the most frequent embarkation point.<br>
For 'Cabin', since there are too many missing values, drop this column.<br>
Verify that there are no more missing values in the processed columns.<br>

```python
# Your Answer: Question 2
df1 = df
df1['Age'] = df1['Age'].fillna(df['Age'].median())
df1.info()

```

```python
df1['Embarked'] = df1['Embarked'].fillna(df['Embarked'].mode()[0])
df1.info()
```

```python
df1 = df1.dropna(axis=1)
df1
```

```python
df1.info() # no missing values
```

## Question 5: Feature Engineering and Relationships
Create a new feature 'FamilySize' by summing 'SibSp' (siblings/spouses) and 'Parch' (parents/children) and adding 1 (for the passenger themselves).<br>
Create a new categorical feature 'Alone' based on 'FamilySize': 1 if FamilySize is 1 (alone), 0 otherwise.<br>
Use Seaborn to create a heatmap of the correlation matrix for numerical features (including 'FamilySize').<br>
Analyze the correlation between 'FamilySize', 'Fare', 'Age', and 'Survived'.

```python
# Your Answer: Question 5

df1['FamilySize'] = df1['SibSp'] + df1['Parch'] + 1
df1
```

```python
df1['Alone'] = (df1['FamilySize']==1).astype(int)
df1
```

```python
num_features = df1[['Survived','Fare','Age','FamilySize']]
sns.heatmap(num_features.corr(),annot=True,cmap='rocket',fmt='.2f')
plt.title('Correlation Heatmap')
```

## Question 6: Advanced Visualization
Create a FacetGrid using Seaborn to visualize the distribution of 'Age' for different 'Pclass' and 'Sex' combinations, with 'Survived' as a hue.<br>
This will involve creating multiple histograms or KDE plots for each combination.

```python
# Your Answer: Question 6

g = sns.FacetGrid(df1, row="Sex", col="Pclass", hue="Survived", margin_titles=True,)
g.map(sns.histplot, "Age", kde=True, bins=20,legend=True)
plt.xlabel("Age")
plt.legend()
plt.ylabel('count')
plt.show()
```
